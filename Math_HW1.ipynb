```python
def y(k,x):
    return(np.cos(k*x))
```

```python
x = np.linspace(-3*np.pi, 3*np.pi, 1000)
y1 = y(1, x)
y2 = y(2, x)
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel("x")
plt.ylabel("y")
```

#### Задание 1.

```python
def len_x (*args):
    summa = 0
    for num in args:
        summa += num**2
    return np.sqrt(summa)
```

```python
len_x(1, 0, -4)
```

```python
def cos_param(x, k, a, b):
    return(k*np.cos(x-a) + b)
```

```python
x = np.linspace(-3*np.pi, 3*np.pi, 1000)
y1 = cos_param(x, 1, 0, 0)
y2 = cos_param(x, 0.5, 1, 1)
y3 = cos_param(x, 5, 2, -5)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.xlabel("x")
plt.ylabel("y")
```
